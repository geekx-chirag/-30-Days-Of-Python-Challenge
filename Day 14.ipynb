# Recursive functions : Functions which recall themselves



# Reversing a string :

```python
def reverse_string(s):
    
    if len(s) <= 1:
        return s
    return reverse_string(s[1:]) + s[0]

print(reverse_string("I love BMW M5 grands sports"))  
```

# Base cases : It prevents infinite recursion

```python
class StringReverser:
    @staticmethod
    def reverse(s):
        return s if len(s) <= 1 else StringReverser.reverse(s[1:]) + s[0]

print(StringReverser.reverse("Let us take an example"))  
```

# Challenge : Calculate factorial recursively 

The image below explains how to calculate factorial in a recursive manner!



# Code :

```python
def factorial(n):
    return 1 if n <= 1 else n * factorial(n - 1)

num = int(input("Enter a non-negative integer: "))

print(f"The factorial of {num} is {factorial(num)}")
```
