# Stacks vs Queue :

| **Aspect**    | **Stack**                   | **Queue**                       |
| ------------- | --------------------------- | ------------------------------- |
| **Structure** | LIFO (Last In, First Out)   | FIFO (First In, First Out)      |
| **Insertion** | Performed at the top (push) | Performed at the rear (enqueue) |
| **Removal**   | From the top (pop)          | From the front (dequeue)        |
| **Use Case**  | Backtracking, undo features | Task scheduling, order handling |
| **Example**   | Browser history, recursion  | Print queue, customer service   |


# Stacks :



# Here i made a list of some random letters

```python
letters = []   # MARVEL 
letters.append('M')
letters.append('A')
letters.append('R')
letters.append('V')
letters.append('E')
letters.append('L')
```

```python
print(letters)
last_item = letters.pop()
print(last_item)
print(letters)
last_item = letters.pop()
print(last_item)
print(letters)
```

# Queues :



| **Operation**  | **Enqueue**                               | **Dequeue**                                     |
| -------------- | ----------------------------------------- | ----------------------------------------------- |
| **Definition** | Adds an element to the back of the queue. | Removes an element from the front of the queue. |
| **Action**     | `queue.append(item)`                      | `queue.pop(0)`|

```python
cars = []  # Cars list
cars.append('Bugatti Chiron Super Sport')
cars.append('Pininfarina Battista')
cars.append('Koenigsegg Jesko')
cars.append('Ferrari SF90 Stradalee')
cars.append('McLaren 720S')
```

# Output :

```python
first_item = cars.pop(0)
print(first_item)
first_item = cars.pop(1)
print(first_item)
print(cars)
```
