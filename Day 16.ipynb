# Day-16: Generators and Iterators 

# Topics - yield, iterators, lazy evaluation

# yield :



```python
def avenger_weapons():
    yield "Thor's Stormbreaker > Mjolnir."
    yield "Captain America Shield."

for weapon in avenger_weapons():
    print(weapon)    # prints 
```

# iterators : repeats itself

```python
# Creating an iterator
mission_log = [
    "The Avengers Infinity War: Avengers Assembled, Wakanda Forever!!",
    "The Avengers Endgame: Dr. Strange points out to Tony Stark for one chnace to defeat Thanos."
]
log_iterator = iter(mission_log)

print(next(log_iterator)) 
print(next(log_iterator))  
```

# lazy evaluation :



```python
def call_avengers():
    team = ["Iron Man", "Captain America", "Thor"]
    for avenger in team:
        yield f"{avenger} ready!!"

for call in call_avengers():
    print(call) # prints
```

# Challenge : Generate the first n Fibonacci numbers with a generator



```python
def avengers_power_levels(n):                   # Its the sum of the previous two numbers
    a, b = 0, 1
    for _ in range(n):
        yield f"Avenger power level: {a}"
        a, b = b, a + b

n = 9  
for power in avengers_power_levels(n):
    print(power)
```
