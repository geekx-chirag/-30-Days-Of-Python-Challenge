# Formatting datetime : you see 146412 (useless?)

```python
# How to fix it simple : by the help of formatting

from datetime import datetime

now = datetime.now().strftime(" %H:%M:%S %m-%d-%y")   
print(now)    # Clean output
```

# Format specifiers for datetime in python :



# Now we can write proper syntax for datetime - 

```python
from datetime import datetime                # Library imported

today = date.today()
print("What is today's date:",today)
print("day:",today.day)
print("month:",today.month)
print("year:",today.year)

print(today.strftime("%A %d-%B-%Y"))
```

# Time difference :

```python
targetdate = datetime(2025, 5, 28, 22, 45)    # 28th May 2025 (Challenge started)
currentdate = datetime(2025, 6, 7, 9, 30)     # 7th June 2025 (Challenge Day 11)

difference = targetdate - currentdate         # Target date has passed away 10 days ago (that's why you see minus sign)

print("Difference:", difference)
print("Days:", difference.days)
print("Seconds:", difference.total_seconds())
```

# Challenge : Calculate the days between two dates

```python
# It same as above :

from datetime import datetime

date1 = datetime(2025, 6, 7)
date2 = datetime(2025, 5, 25)

difference = date1 - date2

print("Days between the two dates are :", difference.days)
```
