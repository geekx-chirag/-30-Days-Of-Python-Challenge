# Topics - Inheritance, polymorphism

#  Inheritance : It is the ability to derive the properties of a parent class into a derived class, just like we study in C++.
#  Types of Inheritance 
#  . Single Inheritance
#  . Multi-level Inheritance
#  . Multiple Inheritance
#  . Hierarchical Inheritance
#  . Hybrid Inheritance



# Single Inheritance :

```python
class Car:
    def drive(self):
        print("Car is driving")

class SportsCar(Car):
    def turbo(self):
        print("Turbo mode on")

c = SportsCar()
c.drive()
```

# Multi-level Inheritance :

```python
class Vehicle:
    def start(self):
        print("Vehicle started")

class Car(Vehicle):
    def drive(self):
        print("Driving car")

class Petrolcar(Car):
    def charge(self):
        print("Refulling tank")

c = Petrolcar()
c.start()
c.drive()
```

# Multiple Inheritance :

```python
class GPS:
    def navigate(self):
        print("Navigation on")

class MusicSystem:
    def play_music(self):
        print("Music on")

class SmartCar(GPS, MusicSystem):
    pass

c = SmartCar()
c.navigate()
c.play_music()
```

# Hierarchical Inheritance :

```python
class Car:
    def start(self):
        print("Car started")

class Sedan(Car):
    def comfort(self):
        print("Comfort mode")

class SUV(Car):
    def offroad(self):
        print("Off-road mode")

c = SUV()
c.start()
```

# Hybrid Inheritance

```python
class Engine:
    def power(self):
        print("Engine power")

class Car(Engine):
    def drive(self):
        print("Driving")

class Feature:
    def autopilot(self):
        print("Autopilot engaged")

class Gwagon(Car, Feature):
    pass

c = Gwagon()
c.power()
c.drive()
c.autopilot()
```

# Polymorphism


```python
class Car:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Drive!")

class Boat:
  def __init__(self, brand, model):
    self.brand = brand
    self.model = model

  def move(self):
    print("Sail!")

car1 = Car("Ford", "Mustang")       # Car object
boat1 = Boat("Ibiza", "Touring 20") # Boat object

for x in (car1, boat1):
  x.move()
```

# Challenge : Extend Car into an ElectricCar subclass with battery capacity

```python
class Car:
    def start(self):
        print("Car started")

class BYD(Car):
    def set_battery(self, capacity):
        self.battery = capacity
    def show_battery(self):
        print("Battery:", self.battery, "kWh")

c = BYD()
c.start()
c.set_battery(95)
c.show_battery()
```
