# Topics -  re module, pattern matching

# re module : In python , "import re" is used to import the regular expressions

```python
import re

text = "I am a proud BMW owner"
pattern = "BMW"

match = re.search(pattern, text)         # It will search about particular word in the given text

if match:
    print("BMW is there")
else:
    print("Nothing found!")
```

# Pattern matching :

```python
import re

text = "Your OTP is 558585585"     # Let us take a random number
pattern = r"\d"

result = re.sub(pattern, "*", text)
print(result)
```

# Challenge : Validate gmail addresses using regex

```python
import re

email = input("Enter your Gmail address: ")

if re.match(r".+@gmail\.com$", email):   # it will match either gmail ends with @gmail.com or not ?!
    print("Valid gmail address")
else:
    print("Not valid gmail address")
```
